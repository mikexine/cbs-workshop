```python
import pandas as pd
```

```python
data = pd.read_csv('./datasets/instagram/export.csv', sep=',', header=0)
```

```python
data
```

```python
data.describe()
```

```python
x_data = data.filter(['comment_count','follows','followed_by','python_faces'], axis=1).fillna(0)
```

```python
y_data = data.filter(['like_count'], axis=1).fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
neigh = KNeighborsRegressor(n_neighbors=15)
```

```python
neigh.fit(x_train, y_train)
```

```python
neigh.score(x_test, y_test)
```

```python
y_predict = neigh.predict(x_test)
```

```python
y_predict = pd.DataFrame(y_predict, y_test.index, columns=['predicted_like_count'])
```

```python
y_obs = pd.concat([y_predict, y_test], axis=1)
```
